```python
import pandas as pd
```

```python
data_wine = pd.read_csv('winemag-data-130k-v2.csv')
data_titanic = pd.read_csv('titanic.csv')
```

```python
data_wine.head()
```

```python
"""
    Renaming
    rename()
    Deixa alterar nomes de index e nomes de colunas
    
    ex.: points para score
    
    RENOMEAR NOME DE COLUNA É MAIS COMUM DO QUE RENOMEAR OS VALORES DO INDEX
"""
```

```python
# alterando nome de alguma coluna específica
data_wine.rename(columns={'points':'score'})
```

```python
# renomeando index
data_wine.rename(index={0:'FirstEntry', 1:'secondEntry'})
```

```python
data_wine.rename_axis('Wines', axis='rows').rename_axis('Fields', axis='columns')
```

```python
"""
    AS VEZES É NECESSÁRIO FAZER A COMBINAÇÃO DE DIFERENTES DATAFRAMES OU SERIES
    O PANDAS POSSUÍ 3 FORMAS DE FAZER ISSO
    
        concat()
        join() --> Faz a junção de diferentes dataframes quando possui index em comum. Se não tiver, não fará a junção
        marge()
    
    
"""
```

```python
# CONCATENANDO DATAFRAME OU SERIES
pd.concat([data_wine, data_titanic])
```
